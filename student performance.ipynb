# boxplot

```python
#shows if the data set has outiers
#shoes the percentile range
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='parental level of education', y='math score', data=df)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education', y='math score', hue='gender', data=df, palette='inferno_r')
```

```python
sns.boxplot(x='test preparation course', y='writing score', data= df)
```

```python
sns.boxplot(x='test preparation course', y='writing score', hue='gender', data= df)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='reading score', y='math score', palette='inferno_r',s=100)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='reading score', y='math score', hue='gender', palette='inferno_r',s=100)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='reading score', y='writing score', hue='gender', palette='inferno_r',s=100)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='lunch', data=df)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='lunch', y='math score', hue='gender', data=df, palette='summer')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='lunch', y='math score', hue='gender', data=df, palette='summer')
plt.legend(bbox_to_anchor=(1.02,1), loc=2, borderaxespad=0)
```
